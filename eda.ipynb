```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn-white')
```

```python
train_data = pd.read_csv('train.csv')
```

```python
train_data.shape
```

```python
train_data.head()
```

```python
#Check for duplicated images
train_data.Image.duplicated().any()
```

```python
distributions = train_data.Id.value_counts()
```

```python
print('Number of unique whales in the dataset: ', distributions.shape[0])
```

```python
top_20 = distributions[0:20]
```

#### 8% of images are photos of new whales. 
#### All whales in the dataset only have a small number of photos. The largest number of photos per whale is 34.

```python
plt.figure(figsize = (20,8))
x = np.arange(len(top_20))
plt.bar(x, top_20.values/train_data.shape[0])
plt.xticks(x, top_20.index, rotation='vertical', fontsize=14)
plt.title('Top Whale IDs by Share of Images', fontsize=16)
plt.ylabel('Share of all images - Log', fontsize=14)
plt.yticks(fontsize=14)
#plt.yscale('log') 
plt.grid()
plt.show()
```
